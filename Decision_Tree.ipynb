we will give some list of depth value and try using for loop to check what will be the accuracy score

```python
depth = [1,2,3,4,5,6,7,8,9]
```

```python
for max_d in depth:
  model = DecisionTreeClassifier(max_depth=max_d,random_state=42)
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  print(f"accuracy score for the model {max_d} is {acc} ")
```

```python
final_tree = DecisionTreeClassifier(max_depth=6,random_state=42)
```

```python
final_tree.fit(x_train,y_train)
y_pred = final_tree.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
#figsize---is a parameter which deals with height and width of a graph
plt.figure(figsize=(12,7))
plot_tree(final_tree,filled=True)
plt.show()
```

```python
#check overfitting
train_pred = final_tree.predict(x_train)
```

```python
accuracy_score(y_train,train_pred)
```

Conclusion:- In our case: 96% vs. 76% is a clear overfit.

With Prepruning Code

```python
#we will store all the decisionTreeClassifier Parameters in a dictionary
parameter={
    "criterion": ['gini','entropy','log_loss'],
    "splitter" : ["best", "random"],
    "max_depth" : [1,2,3,4,5,6],
    "max_features" : ["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
```

```python
#parameters of gridsearcv==estimator, param_grid, *, scoring=None, n_jobs=None, refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs',
#error_score=nan, return_train_score=False

cv = GridSearchCV(tree_model,param_grid= parameter,cv = 5,scoring='accuracy')
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_ #so it has selected criterion = gini,max_depth': 4,'max_features': 'sqrt','splitter': 'best'}
```

```python
y_pred = cv.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
