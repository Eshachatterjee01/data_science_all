```python
#we will delete column player.x and player id..this colimns are note helpfull for machine learning
data.drop(columns = ['Player.x','Player_ID'],inplace = True)
```

Encoding the categorical columns

```python
#first see where its object datatype present in which column
data.select_dtypes('object')
```

```python
obj_cols = data.select_dtypes('object').columns
```

```python
obj_cols
```

```python
#converting object into numeric
le = LabelEncoder()
for col in obj_cols:
  data[col] = le.fit_transform(data[col])
```

```python
data[col].dtype
```

```python
#check for data imbalance
data['Play'].value_counts() ##heavily imbalance
```

Machine Learning Process

```python
x = data.drop(columns = 'Play')
y = data['Play']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

Mean centering/Scalling/Standardization of the data

```python
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

**Now we are goinh to apply simple Logistic Regression on the data**

```python
log_reg = LogisticRegression()
log_reg.fit(x_train_scale,y_train)
```

```python
y_pred = log_reg.predict(x_test_scale)
```

```python
accuracy_score(y_test,y_pred)  ##accuracy_score is unreliable in the case of imbalance data, so we don't trust it
```

```python
#instead i will use roc_auc_score
roc_auc_score(y_test,y_pred)
```

To check for the overfitting

```python
y_pred_train = log_reg.predict(x_train_scale)
```

```python
roc_auc_score(y_train,y_pred_train)
```

**Now we will transform the data with PCA and check the performance of the model using the PCA transformed data**


```python
from sklearn.decomposition import PCA
```

we can mention n_component = 20 ,means how many column or principal component you want that.. but here i used 0.95 that means when i am extracting the information of these 42 columns and creating new columns 95% information has to be extracted properly and 5 % we can loose .based on 95% of information extraction pca will decide the column count..

```python
pca = PCA(n_components=0.95)
```

```python
x_train_pca = pca.fit_transform(x_train_scale)
x_test_pca = pca.transform(x_test_scale)
```

```python
x_train_pca.shape #from 41 column become 21
```

Apply the logistic Regression on the pca_transformed data

```python
log_reg2 = LogisticRegression()
```

```python
log_reg2.fit(x_train_pca,y_train)
```

```python
y_pred_pca = log_reg2.predict(x_test_pca)
```

```python
roc_auc_score(y_test,y_pred_pca)
```

```python
y_pred_train = log_reg2.predict(x_train_pca)
```

```python
roc_auc_score(y_train,y_pred_train)
```

Disclaimer:-  after doing pca transformation also still overfitting.training accuracy 0.911 and test accuracy is 0.76. actually logistic regression,linear regression,svm all of these are not suitable for imbalance dataset..most suitable algorithms are tree based (DT,RF,GradientBoost,XGBoost)..because linear,logistic they are not able to understand patterns properly they are doing overlearning in this case.
