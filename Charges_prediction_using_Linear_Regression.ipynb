```python
#remove or replace null value
col = list(df.columns)
col
```

```python
for i in col:
  if df[i].dtype == "object":
    df[i] = df[i].fillna(df[i].mode()[0])
  else:
    df[i] = df[i].fillna(df[i].mean())
```
